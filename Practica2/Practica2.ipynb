## Carga del archivo de datos

La carga del archivo `datos.csv` se realiza mediante el uso de la biblioteca `pandas`, una herramienta esencial en el análisis de datos en Python.

El proceso de carga del archivo implica leer el contenido del archivo CSV y convertirlo en un DataFrame, lo que permite acceder y trabajar con los datos de manera eficiente. La función `read_csv()` de pandas se encarga de este proceso, permitiendo especificar diversas opciones de configuración según las necesidades del usuario, como el delimitador de campos, los tipos de datos de las columnas, entre otros.


```python
import pandas as pd

dataframe = pd.read_csv('datos.csv')

dataframe
```

## Proceso de Limpieza de Datos

1. **Eliminar Duplicados:**
   Se eliminan las filas duplicadas del DataFrame para garantizar que cada curso esté representado solo una vez.

2. **Limpiar Valores Nulos:**
   Se definen dos funciones de limpieza, una para columnas numéricas y otra para columnas de texto. Las columnas numéricas se llenan con ceros, mientras que las columnas de texto se llenan con cadenas vacías.

3. **Aplicar Funciones de Limpieza:**
   Las funciones de limpieza se aplican a las columnas correspondientes del DataFrame, garantizando que no haya valores nulos en el conjunto de datos.

4. **Imprimir el DataFrame Resultante:**
   Se imprime el DataFrame resultante después de la limpieza, lo que muestra el conjunto de datos listo para su análisis posterior.

Este proceso de limpieza asegura que el DataFrame esté libre de duplicados, lo que facilita su análisis y procesamiento posterior.


```python
# Para eliminar los duplicados
dataframe = dataframe.drop_duplicates()

# Función para limpiar los valores nulos en columnas numéricas
def limpiar_columna_numerica(column):
    return column.fillna(0)

# Función para limpiar los valores nulos en columnas de texto
def limpiar_columna_texto(column):
    return column.fillna('')

# Aplicar funciones de limpieza a columnas numéricas y de texto
columnas_numericas = dataframe.select_dtypes(include=['float64', 'int64']).columns
columnas_texto = dataframe.select_dtypes(include=['object']).columns

dataframe[columnas_numericas] = dataframe[columnas_numericas].apply(limpiar_columna_numerica)
dataframe[columnas_texto] = dataframe[columnas_texto].apply(limpiar_columna_texto)

# Imprimir el DataFrame resultante
dataframe
```

## Cálculo del promedio de calificaciones por curso

El código proporcionado calcula el promedio de calificaciones para cada curso en el conjunto de datos. Esto se logra mediante el uso del método `groupby()` de pandas, que agrupa los datos por el título del curso y luego aplica la función `mean()` para calcular el promedio de las calificaciones de cada grupo.

Una vez calculados los promedios de calificaciones para cada curso, se imprimen los resultados utilizando la función `print()`. Esto proporciona una lista de los cursos junto con sus respectivos promedios de calificaciones, lo que permite una rápida inspección de cómo se distribuyen las calificaciones entre los diferentes cursos.


```python
# Calcular el promedio de calificaciones para cada curso
promedio_por_curso = dataframe.groupby('Course Title')['Rating'].mean()

# Imprimir el promedio de calificaciones para cada curso
print(promedio_por_curso)
```

## Cálculo del mayor y menor rating

El código proporcionado realiza varias operaciones relacionadas con la columna 'Rating' en el conjunto de datos:

1. **Conversión a números y manejo de valores no válidos:** La columna 'Rating' se convierte a valores numéricos utilizando la función `pd.to_numeric()`. Se especifica el argumento `errors='coerce'` para que los valores no válidos se conviertan en NaN (Not a Number).
2. **Eliminación de filas con valores no válidos:** Se eliminan las filas que contienen NaN en la columna 'Rating' utilizando el método `dropna()` de pandas con el argumento `subset=['Rating']`.
3. **Cálculo del curso con el mayor y menor rating:** Se calculan los cursos con el mayor y menor rating utilizando los métodos `idxmax()` y `idxmin()` respectivamente. Estos métodos devuelven los índices de las filas que contienen los valores máximos y mínimos de la columna 'Rating'.
4. **Impresión de resultados:** Se imprimen en pantalla el nombre y el rating del curso con el mayor y menor rating respectivamente, utilizando la función `print()` y f-strings para formatear la salida de manera legible.



```python
# Convertir la columna 'Rating' a números y establecer los valores no válidos como NaN
dataframe['Rating'] = pd.to_numeric(dataframe['Rating'], errors='coerce')

# Eliminar filas con valores de rating no válidos (NaN)
dataframe = dataframe.dropna(subset=['Rating'])

# Calcular el curso con el mayor rating
curso_mayor_rating = dataframe.loc[dataframe['Rating'].idxmax()]

# Calcular el curso con el menor rating
curso_menor_rating = dataframe.loc[dataframe['Rating'].idxmin()]

# Imprimir solo el nombre del curso y su respectivo rating
print("Curso con el mayor rating:")
print(f"Nombre del curso: {curso_mayor_rating['Course Title']}, Rating: {curso_mayor_rating['Rating']}")
print("\nCurso con el menor rating:")
print(f"Nombre del curso: {curso_menor_rating['Course Title']}, Rating: {curso_menor_rating['Rating']}")
```

## Cálculo de porcentaje de cursos flexibles

El código proporcionado realiza un análisis sobre los cursos con horario flexible en el conjunto de datos:

1. **Cálculo del número total de cursos:** Se calcula el número total de cursos en el conjunto de datos original utilizando la función `len(dataframe)`.

3. **Cálculo del número de cursos con horario flexible:** Se cuenta el número de cursos filtrados en el paso anterior utilizando la función `len(cursos_horario_flexible)`.

### Conversiones de Reviews y Horas a formato estándar

El código proporcionado realiza las siguientes tareas:

1. **Extracción del recuento de revisiones:**
   - Se define una función llamada `extraer_review_recuento()` que toma un texto como entrada y busca el recuento de revisiones dentro de ese texto.
   - La función convierte el texto a minúsculas para facilitar la búsqueda y divide el texto en palabras.
   - Luego, busca la palabra 'reviews' en la lista de palabras y verifica si la palabra anterior contiene solo dígitos. Si es así, extrae ese número como el recuento de revisiones.
   - Se aplica esta función a la columna 'Review' del DataFrame para extraer y convertir los recuentos de revisiones en números enteros.
   - Se filtran las filas donde el recuento de revisiones no es None.

2. **Conversión de duraciones a horas:**
   - Se define una función llamada `convertir_duracion_a_horas()` que toma un texto que representa la duración de un curso y lo convierte en horas totales.
   - El texto se convierte a minúsculas para facilitar el análisis y se divide en palabras.
   - Se define un diccionario que mapea unidades de tiempo a sus equivalentes en horas (por ejemplo, 'hour' a 1, 'days' a 24, etc.).
   - Se itera sobre las palabras en el texto y se calcula el total de horas sumando las horas correspondientes a cada número seguido de una unidad de tiempo reconocida.
   - Se aplica esta función a la columna 'Duration' del DataFrame para convertir las duraciones en horas totales.
   - Se filtran las filas donde la duración en horas no es None.


```python
# Función para extraer el recuento de revisiones
def extraer_review_recuento(text):
    # Convertir el texto a minúsculas para facilitar la búsqueda
    text = text.lower()
    # Separar el texto en palabras
    palabras = text.split()
    # Buscar el índice de 'reviews' en las palabras
    if 'reviews' in palabras:
        index = palabras.index('reviews')
        # Si la palabra anterior a 'reviews' contiene solo dígitos, extraer ese número
        if index > 0:
            prev_palabra = palabras[index - 1]
            # Intentar eliminar los caracteres no numéricos y convertir a entero
            cleaned_number = ''.join(filter(str.isdigit, prev_palabra))
            if cleaned_number.isdigit():
                return int(cleaned_number)
    return None

# Aplicar la función para extraer el recuento de revisiones y convertirlo en números enteros
dataframe['Review Count'] = dataframe['Review'].apply(lambda x: extraer_review_recuento(str(x)))

# Filtrar las filas donde 'Review Count' no es None
reviews_numericos = dataframe[dataframe['Review Count'].notnull()]

### Convertir la duración de los cursos a horas
def convertir_duracion_a_horas(text):
    # Convertir el texto a minúsculas para facilitar el análisis
    text = text.lower()
    
    # Definir un diccionario de conversiones de unidades de tiempo a horas
    unidades_tiempo = {
        'hour': 1,
        'hours': 1,
        'minute': 1 / 60,
        'minutes': 1 / 60,
        'day': 24,
        'days': 24,
        'week': 24 * 7,
        'weeks': 24 * 7,
        'month': 24 * 30,
        'months': 24 * 30
    }
    
    # Dividir el texto en palabras
    palabras = text.split()
    
    horas_totales = 0
    
    # Iterar sobre las palabras y calcular el total de horas
    for i in range(len(palabras)):
        palabra = palabras[i]
        if palabra.isdigit():
            # Si la palabra actual es un número, buscar la unidad de tiempo siguiente
            if i + 1 < len(palabras) and palabras[i + 1] in unidades_tiempo:
                unit = palabras[i + 1]
                # Calcular las horas correspondientes y sumarlas al total
                horas_totales += int(palabra) * unidades_tiempo[unit]
                
    return horas_totales

# Aplicar la función para convertir la duración a horas
dataframe['Duration (hours)'] = dataframe['Duration'].apply(convertir_duracion_a_horas)

# Filtrar las filas donde 'Review Count' no es None
horas_numericos = dataframe[dataframe['Duration (hours)'].notnull()]
```

### Distribución de Duraciones de Cursos

El código proporcionado genera un histograma para visualizar la distribución de las duraciones de los cursos, con intervalos de 1000 horas en el eje x:

1. **Cálculo de los límites del eje x:**
   - Se calculan los límites del eje x redondeando hacia abajo el valor mínimo de la duración del curso al múltiplo más cercano de 1000 y redondeando hacia arriba el valor máximo al múltiplo más cercano de 1000 más uno.
   - Estos límites se utilizan para establecer la cantidad de intervalos en función de los nuevos límites.

2. **Creación del histograma:**
   - Se utiliza `plt.hist()` para crear el histograma de la distribución de las duraciones de los cursos.
   - Se especifica el número de intervalos (`num_bins`) y el rango de valores en el eje x utilizando los límites calculados anteriormente.
   - Se asignan colores automáticos a cada barra del histograma.
   
3. **Configuración de aspectos visuales:**
   - Se establece el color de fondo del área de trazado y se configuran el título y las etiquetas de los ejes para una mejor visualización.
   - Se ajustan los límites del eje y para omitir los valores iguales a cero y se establecen los ticks del eje x a intervalos de 1000 horas.

4. **Agregado de etiquetas de datos:**
   - Se agrega el número exacto de cursos en cada barra del histograma para una mejor comprensión de la distribución de datos.
   
5. **Mostrar el histograma:**
   - Se utiliza `plt.show()` para mostrar el histograma generado.


```python
# Calcular los límites del eje x con intervalos de 1000
xmin = 1000 * (dataframe['Duration (hours)'].min() // 1000) 
xmax = 1000 * (dataframe['Duration (hours)'].max() // 1000 + 1) 

# Calcular la cantidad de intervalos basados en los nuevos límites
num_bins = int((xmax - xmin) / 1000)

# Crear el histograma de la distribución de las duraciones de los cursos con intervalos de 1000
plt.figure(figsize=(10, 6), facecolor='lightblue')
n, bins, patches = plt.hist(dataframe['Duration (hours)'], bins=num_bins, range=(xmin, xmax), color='skyblue', edgecolor='black')

# Asignar colores automáticos a cada barra 
colores = plt.cm.tab10(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colores):
    patch.set_facecolor(color)

# Cambiar el color del fondo del área de trazado
plt.gca().set_facecolor('lightgrey')

# Configurar el título y las etiquetas de los ejes con ediciones
plt.title('Distribución de Duraciones de Cursos', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Duración del Curso (horas)', fontsize=12, color='darkgreen')
plt.ylabel('Número de Cursos', fontsize=12, color='darkred')

# Establecer los límites del eje y para omitir los valores iguales a cero
min_freq = min(n[n.nonzero()])
plt.ylim(min_freq, max(n) * 1.1)

# Establecer los ticks del eje x a intervalos de 1000
plt.xticks(np.arange(xmin, xmax+1, 1000))

# Agregar el número exacto de datos en cada barra
for bar, data, left, right in zip(patches, n, bins[:-1], bins[1:]):
    if data != 0:  # Omitir valores iguales a cero
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), int(data) , ha='center', va='bottom', color='black')

# Agregar una cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el histograma
plt.show()
```

#### - El histograma muestra la distribución de las duraciones de los cursos en Coursera. La mayoría de los cursos tienen una duración entre 0 y 1000 horas.

   - Se especifican las etiquetas de los niveles de dificultad en el eje x y se muestran las líneas medias y los puntos de media.

5. **Personalización visual:**
   - Se asignan colores personalizados a las cajas correspondientes a cada nivel de dificultad.
   - Se personalizan los bigotes y la línea de la mediana para una mejor visualización.

6. **Configuración de aspectos visuales:**
   - Se establece el color de fondo del área de trazado y se configuran el título y las etiquetas de los ejes.

7. **Mostrar el gráfico:**
   - Se utiliza `plt.show()` para mostrar el gráfico de cajas generado.


```python
# Eliminar filas con valores nulos en las columnas 'Rating' y 'Level'
dataframe = dataframe.dropna(subset=['Rating', 'Level'])

# Filtrar los cursos que contienen las palabras clave "beginner level", "intermediate level" y "advanced level"
niveles_seleccionados = ['beginner level', 'intermediate level', 'advanced level']
dataframe_filtrado = dataframe[dataframe['Level'].isin(niveles_seleccionados)]

# Obtener las calificaciones por nivel de dificultad
calificaciones_por_nivel = [dataframe_filtrado[dataframe_filtrado['Level'] == nivel]['Rating'] for nivel in niveles_seleccionados]

# Crear el gráfico de cajas
plt.figure(figsize=(10, 6), facecolor='lightblue')
box_plot = plt.boxplot(calificaciones_por_nivel, labels=niveles_seleccionados, showmeans=True, meanline=True, patch_artist=True)

# Colores personalizados para las cajas
colores = ['blue', 'red', 'green']

# Personalizar el color de las cajas
for box, color in zip(box_plot['boxes'], colores):
    box.set_facecolor(color)

# Personalizar los bigotes
plt.setp(box_plot['whiskers'], color='black', linestyle='--')

# Personalizar la línea de la mediana
plt.setp(box_plot['medians'], color='black', linewidth=1.5)

# Cambiar el color del fondo del área de trazado
plt.gca().set_facecolor('lightgrey')

# Configurar el título y las etiquetas de los ejes con ediciones
plt.title('Distribución de Calificaciones de los Cursos', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Nivel de Dificultad', fontsize=12, color='darkgreen')
plt.ylabel('Calificación', fontsize=12, color='darkred')

# Mostrar el gráfico
plt.show()
```
