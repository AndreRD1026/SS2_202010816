#### Carga de los datos
##### Se hace uso la líbreria pandas

```python
import pandas as pd

dataframe = pd.read_csv('datos.csv')

dataframe
```

#### Proceso para la limpieza de los datos

```python
# Para eliminar los duplicados
dataframe = dataframe.drop_duplicates().dropna(subset=['Level'])

# Eliminar duplicados y filas con valores nulos en la columna 'Keyword'
dataframe = dataframe.drop_duplicates().dropna(subset=['Keyword'])

# Función para limpiar los valores nulos en columnas numéricas
def limpiar_columna_numerica(column):
    return column.fillna(0)

# Función para limpiar los valores nulos en columnas de texto
def limpiar_columna_texto(column):
    return column.fillna('')

# Aplicar funciones de limpieza a columnas numéricas y de texto
columnas_numericas = dataframe.select_dtypes(include=['float64', 'int64']).columns
columnas_texto = dataframe.select_dtypes(include=['object']).columns

dataframe[columnas_numericas] = dataframe[columnas_numericas].apply(limpiar_columna_numerica)
dataframe[columnas_texto] = dataframe[columnas_texto].apply(limpiar_columna_texto)

# Imprimir el DataFrame resultante
dataframe
```

#### Cálculo del promedio de calificaciones para cada curso

```python
# Calcular el promedio de calificaciones para cada curso
promedio_por_curso = dataframe.groupby('Course Title')['Rating'].mean()

# Imprimir el promedio de calificaciones para cada curso
print(promedio_por_curso)
```

#### Cálculo del mayor y menor rating

```python
# Calcular el curso con el mayor rating
curso_mayor_rating = dataframe.loc[dataframe['Rating'].idxmax()]

# Calcular el curso con el menor rating
curso_menor_rating = dataframe.loc[dataframe['Rating'].idxmin()]

# Imprimir solo el nombre del curso y su respectivo rating
print("Curso con el mayor rating:")
print(f"Nombre del curso: {curso_mayor_rating['Course Title']}, Rating: {curso_mayor_rating['Rating']}")
print("\nCurso con el menor rating:")
print(f"Nombre del curso: {curso_menor_rating['Course Title']}, Rating: {curso_menor_rating['Rating']}")
```

#### Cálculo de porcentaje de cursos flexibles

```python
# Calcular el número total de cursos en el conjunto de datos original
total_cursos = len(dataframe)

# Filtrar los cursos con horario flexible
cursos_horario_flexible = dataframe[dataframe['Schedule'] == 'Flexible schedule']

# Calcular el número de cursos con horario flexible
num_cursos_horario_flexible = len(cursos_horario_flexible)

# Calcular el porcentaje de cursos con horario flexible respecto al total de cursos
porcentaje_cursos_horario_flexible = (num_cursos_horario_flexible / total_cursos) * 100

# Redondear el porcentaje a dos decimales
porcentaje_cursos_horario_flexible = round(porcentaje_cursos_horario_flexible, 2)

# Imprimir el resultado
print(f"Porcentaje de cursos con horario flexible: {porcentaje_cursos_horario_flexible}%")
```

#### Gráfica 1

```python
import matplotlib.pyplot as plt

# Contar el número de cursos en cada nivel de dificultad
num_cursos_por_nivel = dataframe['Level'].value_counts()

# Crear la gráfica de barras
plt.bar(num_cursos_por_nivel.index, num_cursos_por_nivel.values)

# Configurar el título y las etiquetas de los ejes
plt.title('Número de Cursos por Nivel de Dificultad')
plt.xlabel('Nivel de Dificultad')
plt.ylabel('Número de Cursos')

# Mostrar la gráfica
plt.show()
```

#### Gráfica 2

```python
# Concatenar todas las listas de keywords en una sola lista
all_keywords = [keyword.strip("[]").replace("'", "").split(", ") for keyword in dataframe['Keyword']]
all_keywords_flat = [keyword for sublist in all_keywords for keyword in sublist]

# Contar la frecuencia de cada keyword
keyword_counts = pd.Series(all_keywords_flat).value_counts()

# Seleccionar las 10 categorías principales
top_categories = keyword_counts.head(10)

# Crear la gráfica de barras horizontal
plt.barh(top_categories.index, top_categories.values)

# Configurar el título y las etiquetas de los ejes
plt.title('Número de Cursos en las Principales Categorías')
plt.xlabel('Número de Cursos')
plt.ylabel('Categoría')

# Mostrar la gráfica
plt.show()
```
