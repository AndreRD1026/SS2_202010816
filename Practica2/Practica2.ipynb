#### Carga de los datos

```python
import pandas as pd

dataframe = pd.read_csv('datos.csv')

dataframe
```

#### Proceso para la limpieza de los datos

```python
# Para eliminar los duplicados
dataframe = dataframe.drop_duplicates()

# Función para limpiar los valores nulos en columnas numéricas
def limpiar_columna_numerica(column):
    return column.fillna(0)

# Función para limpiar los valores nulos en columnas de texto
def limpiar_columna_texto(column):
    return column.fillna('')

# Aplicar funciones de limpieza a columnas numéricas y de texto
columnas_numericas = dataframe.select_dtypes(include=['float64', 'int64']).columns
columnas_texto = dataframe.select_dtypes(include=['object']).columns

dataframe[columnas_numericas] = dataframe[columnas_numericas].apply(limpiar_columna_numerica)
dataframe[columnas_texto] = dataframe[columnas_texto].apply(limpiar_columna_texto)

# Imprimir el DataFrame resultante
dataframe
```

#### Cálculo del promedio de calificaciones para cada curso

```python
# Calcular el promedio de calificaciones para cada curso
promedio_por_curso = dataframe.groupby('Course Title')['Rating'].mean()

# Imprimir el promedio de calificaciones para cada curso
print(promedio_por_curso)
```

#### Cálculo del mayor y menor rating

```python
# Calcular el curso con el mayor rating
curso_mayor_rating = dataframe.loc[dataframe['Rating'].idxmax()]

# Calcular el curso con el menor rating
curso_menor_rating = dataframe.loc[dataframe['Rating'].idxmin()]

# Imprimir solo el nombre del curso y su respectivo rating
print("Curso con el mayor rating:")
print(f"Nombre del curso: {curso_mayor_rating['Course Title']}, Rating: {curso_mayor_rating['Rating']}")
print("\nCurso con el menor rating:")
print(f"Nombre del curso: {curso_menor_rating['Course Title']}, Rating: {curso_menor_rating['Rating']}")
```
