#### Carga de los datos
##### Se hace uso la líbreria pandas

```python
import pandas as pd

dataframe = pd.read_csv('datos.csv')

dataframe
```

#### Proceso para la limpieza de los datos

```python
# Para eliminar los duplicados
dataframe = dataframe.drop_duplicates().dropna(subset=['Level'])

# Función para limpiar los valores nulos en columnas numéricas
def limpiar_columna_numerica(column):
    return column.fillna(0)

# Función para limpiar los valores nulos en columnas de texto
def limpiar_columna_texto(column):
    return column.fillna('')

# Aplicar funciones de limpieza a columnas numéricas y de texto
columnas_numericas = dataframe.select_dtypes(include=['float64', 'int64']).columns
columnas_texto = dataframe.select_dtypes(include=['object']).columns

dataframe[columnas_numericas] = dataframe[columnas_numericas].apply(limpiar_columna_numerica)
dataframe[columnas_texto] = dataframe[columnas_texto].apply(limpiar_columna_texto)

# Imprimir el DataFrame resultante
dataframe
```

#### Cálculo del promedio de calificaciones para cada curso

```python
# Calcular el promedio de calificaciones para cada curso
promedio_por_curso = dataframe.groupby('Course Title')['Rating'].mean()

# Imprimir el promedio de calificaciones para cada curso
print(promedio_por_curso)
```

#### Cálculo del mayor y menor rating

```python
# Convertir la columna 'Rating' a números y establecer los valores no válidos como NaN
dataframe['Rating'] = pd.to_numeric(dataframe['Rating'], errors='coerce')

# Eliminar filas con valores de rating no válidos (NaN)
dataframe = dataframe.dropna(subset=['Rating'])

# Calcular el curso con el mayor rating
curso_mayor_rating = dataframe.loc[dataframe['Rating'].idxmax()]

# Calcular el curso con el menor rating
curso_menor_rating = dataframe.loc[dataframe['Rating'].idxmin()]

# Imprimir solo el nombre del curso y su respectivo rating
print("Curso con el mayor rating:")
print(f"Nombre del curso: {curso_mayor_rating['Course Title']}, Rating: {curso_mayor_rating['Rating']}")
print("\nCurso con el menor rating:")
print(f"Nombre del curso: {curso_menor_rating['Course Title']}, Rating: {curso_menor_rating['Rating']}")
```

#### Cálculo de porcentaje de cursos flexibles

```python
# Calcular el número total de cursos en el conjunto de datos original
total_cursos = len(dataframe)

# Filtrar los cursos con horario flexible
cursos_horario_flexible = dataframe[dataframe['Schedule'] == 'Flexible schedule']

# Calcular el número de cursos con horario flexible
num_cursos_horario_flexible = len(cursos_horario_flexible)

# Calcular el porcentaje de cursos con horario flexible respecto al total de cursos
porcentaje_cursos_horario_flexible = (num_cursos_horario_flexible / total_cursos) * 100

# Redondear el porcentaje a dos decimales
porcentaje_cursos_horario_flexible = round(porcentaje_cursos_horario_flexible, 2)

# Imprimir el resultado
print(f"Porcentaje de cursos con horario flexible: {porcentaje_cursos_horario_flexible}%")
```

#### Número de cursos en cada nivel de dificultad

```python
import matplotlib.pyplot as plt

# Convertir el texto de la columna 'Level' a minúsculas
dataframe['Level'] = dataframe['Level'].str.lower()

# Filtrar los cursos que contienen las palabras clave "beginner level", "intermediate level" y "advanced level"
dataframe_filtrado = dataframe[dataframe['Level'].isin(['beginner level', 'intermediate level', 'advanced level'])]

# Contar el número de cursos en cada nivel de dificultad
num_cursos_por_nivel = dataframe_filtrado['Level'].value_counts()

# Colores para cada nivel de dificultad
colores = {'beginner level': 'blue', 'intermediate level': 'green', 'advanced level': 'red'}

# Crear la gráfica de barras
plt.figure(figsize=(10, 6), facecolor='lightblue')

for nivel, num_cursos in num_cursos_por_nivel.items():
    barra = plt.bar(nivel, num_cursos, color=colores.get(nivel, 'grey'))
    plt.text(nivel, num_cursos, str(num_cursos), ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')

# Cambiar el color del fondo
plt.gca().set_facecolor('lightgrey')

# Configurar el título y las etiquetas de los ejes con ediciones
plt.title('Número de Cursos por Nivel de Dificultad', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Nivel de Dificultad', fontsize=12, color='darkgreen')
plt.ylabel('Número de Cursos', fontsize=12, color='darkred')

# Mostrar la gráfica
plt.show()
```

#### Número de cursos en las principales categorías

```python
import numpy as np

# Concatenar todas las listas de keywords en una sola lista
all_keywords = [keyword.strip("[]").replace("'", "").split(", ") for keyword in dataframe['Keyword']]
all_keywords_flat = [keyword for sublist in all_keywords for keyword in sublist]

# Contar la frecuencia de cada keyword
keyword_counts = pd.Series(all_keywords_flat).value_counts()

# Seleccionar las 10 categorías principales
top_categories = keyword_counts.head(10)

# Crear una paleta de colores automática
colores = plt.cm.tab10(np.arange(len(top_categories)))

# Crear la gráfica de barras horizontal
plt.figure(figsize=(10, 6), facecolor='lightblue')

for i, (categoria, num_cursos) in enumerate(zip(top_categories.index, top_categories.values)):
    barra = plt.barh(categoria, num_cursos, color=colores[i])
    plt.text(num_cursos, i, str(num_cursos), ha='left', va='center', fontsize=10, color='black', fontweight='bold')

# Cambiar el color del fondo
plt.gca().set_facecolor('lightgrey')

# Configurar el título y las etiquetas de los ejes con ediciones
plt.title('Número de Cursos en las Principales Categorías', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Número de Cursos', fontsize=12, color='darkgreen')
plt.ylabel('Categoría', fontsize=12, color='darkred')

# Mostrar la gráfica
plt.show()
```

#### Conversiones de Reviews y Horas a formato estándar

```python
# Función para extraer el recuento de revisiones
def extraer_review_recuento(text):
    # Convertir el texto a minúsculas para facilitar la búsqueda
    text = text.lower()
    # Separar el texto en palabras
    palabras = text.split()
    # Buscar el índice de 'reviews' en las palabras
    if 'reviews' in palabras:
        index = palabras.index('reviews')
        # Si la palabra anterior a 'reviews' contiene solo dígitos, extraer ese número
        if index > 0:
            prev_palabra = palabras[index - 1]
            # Intentar eliminar los caracteres no numéricos y convertir a entero
            cleaned_number = ''.join(filter(str.isdigit, prev_palabra))
            if cleaned_number.isdigit():
                return int(cleaned_number)
    return None

# Aplicar la función para extraer el recuento de revisiones y convertirlo en números enteros
dataframe['Review Count'] = dataframe['Review'].apply(lambda x: extraer_review_recuento(str(x)))

# Filtrar las filas donde 'Review Count' no es None
reviews_numericos = dataframe[dataframe['Review Count'].notnull()]

# Imprimir los valores de 'Review Count' que contienen solo números enteros
print(reviews_numericos['Review Count'])


### Otro

def convert_duration_to_hours(text):
    # Convertir el texto a minúsculas para facilitar el análisis
    text = text.lower()
    
    # Definir un diccionario de conversiones de unidades de tiempo a horas
    time_units = {
        'hour': 1,
        'hours': 1,
        'minute': 1 / 60,
        'minutes': 1 / 60,
        'day': 24,
        'days': 24,
        'week': 24 * 7,
        'weeks': 24 * 7,
        'month': 24 * 30,
        'months': 24 * 30
    }
    
    # Dividir el texto en palabras
    palabras = text.split()
    
    horas_totales = 0
    
    # Iterar sobre las palabras y calcular el total de horas
    for i in range(len(palabras)):
        palabra = palabras[i]
        if palabra.isdigit():
            # Si la palabra actual es un número, buscar la unidad de tiempo siguiente
            if i + 1 < len(palabras) and palabras[i + 1] in time_units:
                unit = palabras[i + 1]
                # Calcular las horas correspondientes y sumarlas al total
                horas_totales += int(palabra) * time_units[unit]
                
    return horas_totales

# Aplicar la función para convertir la duración a horas
dataframe['Duration (hours)'] = dataframe['Duration'].apply(convert_duration_to_hours)

# Imprimir los resultados
# Filtrar las filas donde 'Review Count' no es None
horas_numericos = dataframe[dataframe['Duration (hours)'].notnull()]

# print(dataframe[['Duration', 'Duration (hours)']])

print(horas_numericos[['Duration', 'Duration (hours)']])
```

#### Gráfica 4

#### Distribución de Duraciones de Cursos

```python
# Crear el histograma de la distribución de las duraciones de los cursos
plt.figure(figsize=(10, 6), facecolor='lightblue')
n, bins, patches = plt.hist(dataframe['Duration (hours)'], bins=20, color='skyblue', edgecolor='black')

# Asignar colores automáticos a cada barra 
colores = plt.cm.tab10(np.linspace(0, 1, len(patches)))
for patch, color in zip(patches, colores):
    patch.set_facecolor(color)

# Cambiar el color del fondo del área de trazado
plt.gca().set_facecolor('lightgrey')

# Configurar el título y las etiquetas de los ejes con ediciones
plt.title('Distribución de Duraciones de Cursos', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Duración del Curso (horas)', fontsize=12, color='darkgreen')
plt.ylabel('Frecuencia', fontsize=12, color='darkred')

# Agregar el número exacto de datos en cada barra
for bar, data in zip(patches, n):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), int(data), ha='center', va='bottom', color='black')

# Agregar una cuadrícula
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el histograma
plt.show()
```

#### Gráfica 5

#### Distribución de las calificaciones por nivel de dificultad

```python
# Eliminar filas con valores nulos en las columnas 'Rating' y 'Level'
dataframe = dataframe.dropna(subset=['Rating', 'Level'])

# Obtener los niveles de dificultad únicos
niveles_dificultad = dataframe['Level'].unique()

# Crear una lista para almacenar las calificaciones por nivel de dificultad
calificaciones_por_nivel = []

# Iterar sobre los niveles de dificultad y obtener las calificaciones correspondientes
for nivel in niveles_dificultad:
    calificaciones = dataframe[dataframe['Level'] == nivel]['Rating']
    calificaciones_por_nivel.append(calificaciones)

# Crear el gráfico de cajas
plt.figure(figsize=(10, 6) , facecolor='lightblue')
box_plot = plt.boxplot(calificaciones_por_nivel, labels=niveles_dificultad, showmeans=True, meanline=True, patch_artist=True)

# Colores personalizados para las cajas
colores = ['blue', 'red', 'green']

# Personalizar el color de las cajas
for box, color in zip(box_plot['boxes'], colores):
    box.set_facecolor(color)

# Personalizar los bigotes
plt.setp(box_plot['whiskers'], color='black', linestyle='--')

# Personalizar la línea de la mediana
plt.setp(box_plot['medians'], color='black', linewidth=1.5)

# Cambiar el color del fondo
plt.gca().set_facecolor('lightgrey')

# Configurar el título y las etiquetas de los ejes con color de fondo
plt.title('Distribución de Calificaciones de los Cursos', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Nivel de Dificultad', fontsize=12, color='darkgreen')
plt.ylabel('Calificación', fontsize=12, color='darkred')

# Mostrar el gráfico
plt.show()
```
